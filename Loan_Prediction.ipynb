SPLIT THE TRAINING DATASET

```python
X_train = train.drop([ 'Loan_Status'], axis=1)
y_train = train['Loan_Status']
```

```python
X_train
```

```python
y_train
```

CONVERT THE CATEGORICAL DATA TO NUMBERS

```python
y_train = y_train.replace({'Y': 1, 'N': 0})
y_train
```

FILL THE NULL VALUES

```python
num = X_train.select_dtypes(include='number').columns
X_train[num] = X_train[num].fillna(X_train[num].mean())
```

```python
categorical_cols = X_train.select_dtypes(include='object').columns
for col in categorical_cols:
    if X_train[col].notnull().any():  # Check if there are any non-null values
        X_train[col] = X_train[col].fillna(X_train[col].mode().iloc[0])
```

```python
X_train.isnull().sum()
```

```python
X_test = test
```

```python
X_test
```

```python
X_test.isnull().sum()
```

```python
numerical_cols = X_train.select_dtypes(include='number').columns
X_test[numerical_cols] = X_test[numerical_cols].fillna(X_train[numerical_cols].mean())
```

```python
categorical_cols = X_train.select_dtypes(include='object').columns
for col in categorical_cols:
    X_test[col].fillna(X_train[col].value_counts().idxmax(), inplace=True)
```

```python
X_test.isnull().sum()
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

ONE HOT ENCODING

```python
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)
```

```python
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)
```

```python
scaler = StandardScaler()
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])
```

USING LOGISTIC REGRESSION TO DO CLASSIFICATION

```python
regressor = LogisticRegression()
regressor.fit(X_train, y_train)
```

PREDICT THE LOAN STATUS

```python
y_pred = regressor.predict(X_test)
```

```python
y_test=y_train
```

CHECKING THE ACCURACY

```python
predictions = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(predictions)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
